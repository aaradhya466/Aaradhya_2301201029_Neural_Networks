```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np

# Sample Data
X_data = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
Y_data = np.array([[0],[1],[1],[0]], dtype=float)

# Placeholders
X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

# Weights & Biases
W1 = tf.Variable(tf.random_normal([2, 4]))
b1 = tf.Variable(tf.zeros([4]))

W2 = tf.Variable(tf.random_normal([4, 1]))
b2 = tf.Variable(tf.zeros([1]))

# Hidden Layer
layer1 = tf.nn.relu(tf.matmul(X, W1) + b1)

# Output Layer
output = tf.sigmoid(tf.matmul(layer1, W2) + b2)

# Loss
loss = tf.reduce_mean(tf.square(output - Y))

# Optimizer
train = tf.train.GradientDescentOptimizer(0.1).minimize(loss)

# Session
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    for i in range(1000):
        sess.run(train, feed_dict={X: X_data, Y: Y_data})

    print("Output:")
    print(sess.run(output, feed_dict={X: X_data}))
```

```python
import numpy as np
```

```python
# sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# derivative
def sigmoid_der(x):
    return x * (1 - x)
```

```python
# input
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[1],[1],[0]])
```

```python
# weights
np.random.seed(1)
W1 = np.random.rand(2,4)
W2 = np.random.rand(4,1)
```

```python
for i in range(10000):

    # forward
    A1 = sigmoid(np.dot(X, W1))
    A2 = sigmoid(np.dot(A1, W2))

    # error
    error = Y - A2

    # backprop
    dA2 = error * sigmoid_der(A2)
    dW2 = np.dot(A1.T, dA2)

    dA1 = np.dot(dA2, W2.T) * sigmoid_der(A1)
    dW1 = np.dot(X.T, dA1)

    # update
    W1 += dW1 * 0.1
    W2 += dW2 * 0.1

print("Output:", A2)
```
